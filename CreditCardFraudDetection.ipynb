# **DATA** **PREPROCESSING**

```python
import pandas as pd
import kagglehub
import os
path = kagglehub.dataset_download("kartik2112/fraud-detection")
csv_file = None
for root, _, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            csv_file = os.path.join(root, file)
            break
    if csv_file:
        break

if csv_file:
    df = pd.read_csv(csv_file)
    print("Successfully loaded the CSV file.")
    display(df.head())
else:
    print("No CSV file found in the downloaded dataset directory.")
df = pd.read_csv(csv_file)
df.head()
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
df["dob"] = pd.to_datetime(df["dob"])

df["transaction_hour"] = df["trans_date_trans_time"].dt.hour
df["age"] = df["trans_date_trans_time"].dt.year - df["dob"].dt.year

df.head()
```

```python
df.drop(["first", "last", "street", "city", "state", "zip", "job", "trans_num"], axis=1, inplace=True)
df.head()
```

Longitude and Latitude are given we can convert it into distances b/w user and merchant

```python
from geopy.distance import geodesic

df = df.dropna(subset=["lat", "long", "merch_lat", "merch_long"])

df["location_diff"] = df.apply(lambda row: geodesic(
    (row["lat"], row["long"]), (row["merch_lat"], row["merch_long"])
).km, axis=1)

df.head()
```

**Encoding Categorical Features**

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

label_encoders = {}
for col in ["category", "merchant", "gender"]:
  le = LabelEncoder()
  df[col] = le.fit_transform(df[col])
  label_encoders[col] = le
```

## **Defining X AND** **Y**

```python
X = df.drop("is_fraud", axis=1)
y = df["is_fraud"]
```

# TRAIN/TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

*Handling Class imblance*

```python
from imblearn.over_sampling import SMOTE

X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_test = X_test.apply(pd.to_numeric, errors='coerce')

sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_train, y_train)
```

```python
scaler = StandardScaler()

X_resampled = scaler.fit_transform(X_resampled)
X_test = scaler.transform(X_test)
```

## Train Random Forest


```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_resampled, y_resampled)
```

# Predict

```python
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]
```

# Evaluation

```python
from sklearn.metrics import classification_report, roc_auc_score

print(classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```
